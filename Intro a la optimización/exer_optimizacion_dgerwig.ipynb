## Vallado de un jardin

Queremos determinar las medidas $x$ e $y$ que crearán un jardín con una superficie máxima utilizando 100 metros de vallado.

![jardin](https://openstax.org/apps/archive/20240603.181933/resources/b628896cf794d3d6e3f4672e01630e4a67ab2005)

Definir las variables:
- x: Largo
- y: Ancho 

La ecuación del perímetro para un jardín rectangular es 2x+2y=100, 
lo que se simplifica a x+y=50.


El área del jardín es A=x⋅y

Usando la relación del perímetro, podemos expresar y=50−x.

Po lo que el área en función de $A(x) = x \cdot (50 - x) = 50x - x^2$


```python
from scipy.optimize import minimize_scalar

# Definimos la función que queremos maximizar (el área del jardín)
def area(x):
    return -(50*x - x**2)  # Ponemos un signo negativo para maximizar

# Usamos minimize_scalar para encontrar el máximo
res = minimize_scalar(area)

x_opt = res.x
y_opt = 50 - x_opt
max_area = x_opt * y_opt

print("Resultados:")
print(f"Largo óptimo del jardín (x): {x_opt:.2f} metros")
print(f"Ancho óptimo del jardín (y): {y_opt:.2f} metros")
print(f"Área máxima del jardín: {max_area:.2f} metros cuadrados")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(0, 50)
y_vals = [area(x)*-1 for x in x_vals]

plt.plot(x_vals, y_vals);
plt.scatter(x_opt, max_area, color='red');
```

# Maximizando ingresos

Los propietarios de una empresa de alquiler de coches han determinado que si cobran a los clientes $p$ euros diarios por el alquiler un coche, donde $50 \le p \le 200$, el número de coches $n$ que alquilan por día puede ser modelado por la función lineal $n(p)=1.000−5p$. Es decir, dependiendo de $p$, $n$ tomará distintos valores para el número de coches que se alquilarán. Esto implica que el rendimiento total será de $n(p) \times p$ para un día dado.

Si cobran 50€ por día o menos, alquilarán todos sus automóviles. Si cobran 200€ por día o más, no alquilarán ningún automóvil. Suponiendo que los propietarios planeen cobrar a los clientes entre 50 diarios y 200 diarios por el alquiler de un coche, ¿cuánto deberían cobrar para maximizar sus ingresos?

Definir la función de ingresos totales I(p)=p⋅(1000−5p)

Es decir, $I(p) = 1000p - 5p^2$


```python
from scipy.optimize import minimize_scalar

# Definimos la función de ingresos totales I(p)
def total_ingresos(p):
    return -(1000*p - 5*p**2)  # Usamos signo negativo para maximizar

# Usamos minimize_scalar para encontrar el máximo de la función
res = minimize_scalar(total_ingresos, bounds=(50, 200))

# El valor óptimo de p que maximiza los ingresos
p_opt = res.x

# Calculamos el número de coches alquilados correspondiente
n_opt = 1000 - 5*p_opt

# Calculamos el ingreso máximo
max_ingresos = n_opt * p_opt

# Mostramos los resultados
print(f"El precio óptimo para maximizar los ingresos es: {p_opt:.2f} € / día")
print(f"Número de coches alquilados: {n_opt:.0f}")
print(f"Ingreso máximo diario: {max_ingresos:.2f} €")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(50, 200)
y_vals = [total_ingresos(x)*-1 for x in x_vals]

plt.plot(x_vals, y_vals);
plt.scatter(p_opt, max_ingresos, color='red');
```
